### <font color='blue'>テキストコーパスの探索</font>

特定のチャンカにマッチする句を、テキストから抜き出すことも可能。<br>

```python
# ライブラリのimport
import nltk
```

```python
# チャンキング文法の構築
cp = nltk.RegexpParser('CHUNK: {<V.*> <TO> <V.*>}')
```

```python
# ブラウンコーパスの読み込み
brown = nltk.corpus.brown
```

```python
for sent in brown.tagged_sents()[:20]:
    tree = cp.parse(sent)
    for subtree in tree.subtrees():
        if subtree.label() == 'CHUNK':print(subtree)
```

### <font color='blue'>チンキング</font>

チャンクから取り除きたいものを定義することが簡単な場合もある。<br>
その場合、チャンクに含まれないトークンの配列をチンクとして定義可能。<br>

以下では、'barked/VBD' 'at/IN' がチンクである。<br>

```python
from IPython.display import Image
Image(filename='IMG/7-6.png', width=600)
```

チンキングはチャンクからトークン列を取り除く作業である。<br>

```python
# チャンク文法とチンキング
grammar = """
    NP:
    {<.*>+}          # 全てをチャンクとしている
    }<VBD|IN>+{      # VBD/INをチンクとしている
"""
```

```python
# チンキングする文章
sentence = [("the", "DT"), ("little", "JJ"), ("yellow", "JJ"),
       ("dog", "NN"), ("barked", "VBD"), ("at", "IN"),  ("the", "DT"), ("cat", "NN")]
```

```python
cp = nltk.RegexpParser(grammar)
```

```python
# 以下ではチンクを取り除いている
print(cp.parse(sentence))
```

```python
cp.parse(sentence)
```

### <font color='blue'>チャンクの表現方法 : タグと木構造</font>

チャンク構造はタグ付けと構文解析の中間の状態が適している。<br>
したがって、タグか木構造のどちらかを使って表すことができる。<br>

もっとも広く普及しているファイルの表現方法はIOBタグというものである。<br>

この手法では、それぞれのトークンは<br>
I(inside)、O(outside)、B(begin)という３つの特殊なチャンクタグのいずれかでタグ付けされる。<br>

チャンクの先頭に存在するトークンにはB。<br>
それに続くチャンク内にあるトークンにはI。<br>
それ以外のトークンにはOというタグが付けられる。<br>

BとIのタグには、B-NP、I-NPというように、チャンクタイプが接尾辞として付けられる。<br>

以下に例を示す。<br>

```python
from IPython.display import Image
Image(filename='IMG/7-7.png', width=600)
```

IBOタグはファイル内でのチャンク構造を表す標準の形になっている。<br>
