### <font color='blue'>Nグラムを超えて</font>

以前にバイグラムを用いて、文章を作成するということを行なった。<br>
しかし、バイグラムを用いて文章を作成しても、意味が通るものはなかった。<br>

以下にその例をあげる。<br>

```python
# 画像の表示
from IPython.display import Image
Image(filename='IMG/8-3.png', width=500)
```

上記の文を見ても、どこが悪いのか分からない。<br>
しかし、直感的に意味をなさない「ワードサラダ」と言うことが分かる。<br>

以下に文法的に上記の文のどこが間違っているかを記述していく。<br>

#### <font color='blue'>「the worst part and clumsy looking」について</font>

上記の文は等位接続であることが分かる。<br>

以下に等位接続の例をあげる。<br>

```python
# 画像の表示
from IPython.display import Image
Image(filename='IMG/8-4.png', width=500)
```

一つ目の文では、2つのNP(名詞句)が接続されて、NPを作成していることが分かる。<br>
二つ目の文では、２つのAP（形容詞句）が接続されて、APを作成していることが分かる。<br>

しかし、等位接続では、NPとAPを接続することはできない。<br>
したがって、「the worst part and clumsy looking」という部分が非文法的であることが分かる。<br>
