「I shot an elephant in my pajamas.」という句に注目する。<br>
以下のように簡単な文法を定義する。<br>

```python
import nltk

# 文法を定義
groucho_grammar = nltk.CFG.fromstring("""
S -> NP VP
PP -> P NP
NP -> Det N | Det N PP | 'I'
VP -> V NP | VP PP
Det -> 'an' | 'my'
N -> 'elephant' | 'pajamas'
V -> 'shot'
P -> 'in'
""")
```

各チャンクの意味について以下の２つのURLが詳しい。<br>
http://phiz.c.u-tokyo.ac.jp/~yatabe/6-syntax1.pdf<br>
http://www.phontron.com/slides/nlp-programming-ja-10-parsing.pdf<br>

この文法のもとで、この文は２つの解釈ができる。<br>
「in my pajamas」という前置詞句が<br>
elephantを記述するのか、shootという行為を記述するのか<br>
の２つである。<br>

```python
# 解釈する文章
sent = ['I', 'shot', 'an', 'elephant', 'in', 'my', 'pajamas']
```

```python
# パーサーの作成
parser = nltk.ChartParser(groucho_grammar)
```

```python
# 構文木の作成
for tree in parser.parse(sent):
    print(tree)
```
