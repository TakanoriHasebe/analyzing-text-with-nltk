### <font color='blue'>文法的なジレンマ</font>

間違った文法で書かれた英文。<br>
New man at the of IMG.<br>

間違った文法で書かれた英文を作るのが簡単であるのと同様に、正しい文法で書かれた英文を作成すすることも簡単。<br>
以下に小さな文から、大きな文を作成する例をあげる。<br>

a. Usain Bolt broke the 100m record.<br>
b. The Jamaica Observer reported that Usain Bolt broke the 100m record.<br>
c. Andre said The Jamaica Observer reported that Usain Bolt broke the 100m record.<br>
d. I think Andre said the Jamaica Observer reported that Usain Bolt broke the 100m record.<br>

文全体をSというシンボルに置き換えられるのら<br>
「Andre said S」や「I think S」のようなパターンが存在する。<br>
他にも「S but S」や「S when S」というテンプレートも作成できる。<br>
