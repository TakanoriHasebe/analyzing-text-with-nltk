上記の最初の３つの処理
<ol>
<li>文分割</li>
<li>単語トークナイザ</li>
<li>品詞タガー</li>
</ol>
を行う関数を以下に示す。

```python
import nltk

def ie_preprocess(document):
    sentences = nltk.sent_tokenize(document) # 文分割
    sentences = [ nltk.word_tokenize(sent) for sent in sentences ] # 単語トークナイザ
    sentences = [nltk.pos_tag(sent) for sent in sentences] # 品詞タガー
```

### <font color='blue'>名詞句チャンキング</font>

名詞句チャンキング、NPチャンキングと呼ばれる、個別の名詞句に対応するチャンクを特定するタスクから始める。<br>

以下にNPチャンクが大括弧で記された Wall Street Journal のテキストを表示する。<br>

```python
Image(filename='IMG/7-3.png', width=700)
```

上記からわかるように、完全な名詞句<br>
「the market for system-management software for Digital's hardware」<br>
よりも<br>
「the market」<br>
というよりシンプルなチャンクになっている。<br>

このことは、NPチャンクが他のNPチャンクを含まないということが１つの理由である。<br>
したがって、名詞句を修飾する前置詞句や従属節は対応するNPチャンクに含まれない。<br>
なぜならほぼ確実に、別の名詞句を含むからである。<br>

NPチャンキングのためにもっとも有用な情報源の１つは、品詞タグである。<br>
品詞タグ付けを行う１つの理由は、NPチャンキングである。<br>

品詞タグ付けが行われた文章を用い、上記のアプローチを以下に示す。

```python
# 品詞タグ付けが行われた文章
sentence = [("the", "DT"), ("little", "JJ"), ("yellow", "JJ"), ("dog", "NN"), ("barked", "VBD"), ("at", "IN"),  ("the", "DT"), ("cat", "NN")]
```

チャンク文法は、文がどのようにチャンキングされるか、というルールで構成されている。<br>

```python
# チャンク文法
# 省略可能な限定詞<DT>に続いて、任意の数の形容詞<JJ>が出現し、最後に名詞<NN>が来る
grammar = "NP: {<DT>?<JJ>*<NN>}"
```

```python
# nltkを用いたチャンキング文法の構築
cp = nltk.RegexpParser(grammar)
```

```python
# 例文を使ってテスト
result = cp.parse(sentence)
```
