```python
# 各単語が文章中で平均何回現れているか
def lexical_diversity(text):
    return len(text) / len(set(text))
```

```python
# %で算出
def percentage(count, total): 
    return 100 * count / total
```

```python
lexical_diversity(text3), lexical_diversity(text5)
```

## <font color='blue'>言語の計算処理:簡単な統計処理</font>

### <font color='blue'>頻度分布</font>

### <font color='blue'>FreqDistを用いて"Moby Dick"から最も頻出する50単語の抜き出し</font>

```python
from collections import OrderedDict

# テキストを渡して、単語の総数を表示
fdist1 = FreqDist(text1)
fdist1 = dict(fdist1)
```

```python
# 出現順に昇順で並び替え
fdist1 = OrderedDict(sorted(fdist1.items(), key=lambda x:x[1], reverse=True))
```

```python
fdist1
```

```python
# 単語をとってくる
vocabulary1 = fdist1.keys()
```

```python
# 最初から50個の単語を表示
list(vocabulary1)[:50]
```

```python
# 'whale'が何番目か
fdist1['whale']
```

```python
# 累積頻度をプロット
FreqDist(fdist1).plot(50, cumulative=True)
```
