```python
# 各単語が文章中で平均何回現れているか
def lexical_diversity(text):
    return len(text) / len(set(text))
```

```python
# %で算出
def percentage(count, total): 
    return 100 * count / total
```

```python
lexical_diversity(text3), lexical_diversity(text5)
```

## <font color='blue'>言語の計算処理:簡単な統計処理</font>

### <font color='blue'>頻度分布</font>

### <font color='blue'>FreqDistを用いて"Moby Dick"から最も頻出する50単語の抜き出し</font>

```python
from collections import OrderedDict

# テキストを渡して、単語の総数を表示
fdist1 = FreqDist(text1)
fdist1 = dict(fdist1)
```

```python
# 出現順に昇順で並び替え
fdist1 = OrderedDict(sorted(fdist1.items(), key=lambda x:x[1], reverse=True))
```

```python
fdist1
```

```python
# 単語をとってくる
vocabulary1 = fdist1.keys()
```

```python
# 最初から50個の単語を表示
list(vocabulary1)[:50]
```

```python
# 'whale'が何番目か
fdist1['whale']
```

```python
# 累積頻度をプロット
FreqDist(fdist1).plot(50, cumulative=True)
```

## <font color='blue'>よりきめ細かい単語選択</font>

```python
# 15文字以上の単語の抜き出し
V = set(text1)
long_words = [w for w in V if len(w) > 15]
sorted(long_words)
```

```python
# コーパスから7文字よりも長く、7回よりも多く出現する全ての単語の抜き出し
fdist5 = FreqDist(text5)
sorted([w for w in set(text5) if len(w) > 7 and fdist5[w] > 7])
```

## <font color='blue'>バイグラムとコロケーション</font>

```python
# バイグラム
import nltk
from nltk import word_tokenize
from nltk.util import ngrams
from collections import Counter
text = "Hi How are you? i am fine and you"
token=nltk.word_tokenize(text)
bigrams=ngrams(token,2)
```

```python
Counter(bigrams)
```

<font color='red'>コロケーションとは、非常に頻繁に共起する一連の単語列を意味する。</font><br>
