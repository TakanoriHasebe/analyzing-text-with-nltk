### <font color='blue'>NLTKで木構造を構築する</font>

```python
import nltk

# 木構造の作成
tree1 = nltk.Tree('NP', ['Alice'])
```

```python
tree1
```

```python
# 木構造の作成
tree2 = nltk.Tree('NP', ['the', 'rabbit'])
```

```python
# 木構造についてみる
tree2
```

以下により大きな木構造を作成する。<br>

```python
# より深い木構造の作成
tree3 = nltk.Tree('VP', ['chased', tree2])
tree4 = nltk.Tree('S', [tree1, tree2])
```

```python
# 木構造
tree4
```

```python
# 木構造の右側をみる
tree4[1]
```

```python
# 何チャンクかをみる
tree4[1].label()
```

```python
# 木構造の最後の葉の部分をみる
tree4.leaves()
```

```python
# 単語
tree4[1][1]
```

```python
# 単語のスペル
tree4[1][1][0], tree4[1][1][1]
```

### <font color='blue'>木の走査</font>

以下に木の走査のための再帰的関数を作成する。<br>

```python
def traverse(t):
    try:
        t.label()
    except AttributeError:
        print(t, end=" ")
    else:
        # Now we know that t.node is defined
        print('(', t.label(), end=" ")
        for child in t:
            traverse(child)
        print(')', end=" ")
```

```python
t = nltk.Tree('(S (NP Alice) (VP chased (NP the rabbit)))')
```
