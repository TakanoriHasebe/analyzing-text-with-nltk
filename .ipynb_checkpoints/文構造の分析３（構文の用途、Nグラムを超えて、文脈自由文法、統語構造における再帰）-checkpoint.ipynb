#### <font color='blue'>「the worst part and clumsy looking」について</font>

上記の文は等位接続であることが分かる。<br>

以下に等位接続の例をあげる。<br>

```python
# 画像の表示
from IPython.display import Image
Image(filename='IMG/8-4.png', width=500)
```

一つ目の文では、2つのNP(名詞句)が接続されて、NPを作成していることが分かる。<br>
二つ目の文では、２つのAP（形容詞句）が接続されて、APを作成していることが分かる。<br>

しかし、等位接続では、NPとAPを接続することはできない。<br>
したがって、「the worst part and clumsy looking」という部分が非文法的であることが分かる。<br>

ここで、句と節の違いについて述べる。<br>
句についてはその内部に< S + V >という構造を持たないのに対し、節についてはその内部に< S + V >という構造をもつ。<br>

### <font color='blue'>文脈自由文法</font>

#### <font color='blue'>簡単な文法</font>

以下に文法を定義し、この文法に沿って適格な文をどのように構文解析するかを示す。<br>

```python
import nltk

# 文法の定義
grammar1 = nltk.CFG.fromstring("""
  S -> NP VP
  VP -> V NP | V NP PP
  PP -> P NP
  V -> "saw" | "ate" | "walked"
  NP -> "John" | "Mary" | "Bob" | Det N | Det N PP
  Det -> "a" | "an" | "the" | "my"
  N -> "man" | "dog" | "cat" | "telescope" | "park"
  P -> "in" | "on" | "by" | "with"
  """)
```

```python
# 構文解析する文章
sent = "Mary saw Bob".split()
```

```python
# 文章を分ける
sent
```

```python
# パーサーの作成
rd_parser = nltk.RecursiveDescentParser(grammar1)
```

```python
# 型について
type(rd_parser), rd_parser
```

```python
# 構文解析の実行
for tree in rd_parser.parse(sent):
    print(tree)
```

上記のように「 | 」で分けると２つの分岐点ができる。<br>
したがって、文章によっては２つの構文木が作成されることがある。<br>

上記のように行うことを「構文的に曖昧」であるという。<br>

```python
grammar2 = nltk.CFG.fromstring("""
  S  -> NP VP
  NP -> Det Nom | PropN
  Nom -> Adj Nom | N
  VP -> V Adj | V NP | V S | V NP PP
  PP -> P NP
  PropN -> 'Buster' | 'Chatterer' | 'Joe'
  Det -> 'the' | 'a'
  N -> 'bear' | 'squirrel' | 'tree' | 'fish' | 'log'
  Adj  -> 'angry' | 'frightened' |  'little' | 'tall'
  V ->  'chased'  | 'saw' | 'said' | 'thought' | 'was' | 'put'
  P -> 'on'
  """)
```
