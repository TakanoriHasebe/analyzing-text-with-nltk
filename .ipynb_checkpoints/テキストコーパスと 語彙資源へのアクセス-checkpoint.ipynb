### <font color='blue'>各テキストについて他の情報を表示する</font>

#### <font color='blue'>それぞれのテキストに対して、単語の平均長、文の平均長、各単語がテキスト中に現れた回数の平均(語彙多様性スコア)</font>

```python
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid))
    num_vocab = len(set([w.lower() for w in gutenberg.words(fileid)]))
    print (int(num_chars/num_words), int(num_words/num_sents), int(num_words/num_vocab), fileid)
```

#### <font color='blue'>法助動詞の用法から、ジャンルにおける違いを比較</font>

```python
from nltk.corpus import brown

categories = brown.categories()
del categories[2]
for c in categories:
    print(c)
    news_text = brown.words(categories = c)
    fdist = nltk.FreqDist([w.lower() for w in news_text])
    modals = ['can', 'could', 'may', 'might', 'must', 'will']
    for m in modals:
        print(m + ':', fdist[m])
    print('')
```

#### <font color='blue'>条件付き頻度分布のサポート機能を用いる</font>

下の結果から<br>
ロマンスのジャンルにおいてもっとも頻繁に現れる法助動詞が「could」なのに対し、ニュースの 場合は「will」であることがわかる。

```python
cfd = nltk.ConditionalFreqDist(
            (genre, word)
            for genre in categories
            for word in brown.words(categories=genre)
)
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditons=genres, samples=modals)
```

### <font color='blue'>語彙資源</font>

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab.difference(english_vocab)
    return sorted(unusual)
```

```python
unusual_words(nltk.corpus.gutenberg.words('austen-sense.txt'))
```

```python
unusual_words(nltk.corpus.nps_chat.words())
```

### <font color='blue'>ストップワード</font>

高頻度で登場する言葉をストップワードという。

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

あるテキスト中にストップワードが含まれない単語の割合を計算する。

```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / len(text)

content_fraction(nltk.corpus.reuters.words())
```

上記のようにストップワードを考慮すると、テキスト中の３分の１を無視できるようになった。

### <font color='blue'>発音辞書</font>

単語だけでなくそれぞれの単語の属性情報を各行に保持した表(スプレッドシート)データ<br>
NLTK は、音声合成装置で利用するために設計された、発音辞書(CMU Pronouncing Dictionary)を含んでいる。<br>

```python
entries = nltk.corpus.cmudict.entries()
```

```python
len(entries)
```

```python
# explorersには２つの発声があるということがわかる。
for entry in entries[39943:39951]:
    print(entry)
```

上記を見ると各エントリは２つの部分から構成されている。<br>
少し複雑なfor文とプログラムにより、詳しく見ることができる。<br>

```python
for word, pron in entries:
    if len(pron) == 3:
        ph1, ph2, ph3 = pron
        if ph1 == 'P' and ph3 == 'T':
            print(word, ph2)
```

上記のプログラムは語彙目録を走査して、3つの音素からなる発音をもつエントリを探す。<br>

下記のプログラムは"nicks"のような発音で終わる単語を全て見つけ出す。<br>

```python
syllable = ['N', 'IH0', 'K', 'S']
[word for word, pron in entries if pron[-4:] == syllable]
```

上記から１つの発音が複数の表記で記される場合があるということに気づく。<br>

発音と表記の他のミスマッチの例を探して見る。<br>

以下は最後がmnで終わり、どちらかしか発音しない。<br>

```python
[w for w, pron in entries if pron[-1] == 'M' and w[-1] == 'n']
```

w[0]が'n'でなく、pron[0]が'N'で、重複を含まない。<br>

```python
sorted(set(w[:2] for w, pron in entries if pron[0] == 'N' and w[0] != 'n'))
```

語彙目録をスキャンして、発音の数字を抽出<br>
特定のアクセントパターンを 持つ語を見つけ出す関数を定義<br>

```python
def stress(pron):
    return [char for phone in pron for char in phone if char.isdigit()]
[w for w, pron in entries if stress(pron) == ['0', '1', '0', '2', '0']]
```

条件付き頻度分布は、単語の対照的な組み合わせを簡単に見つけることもできる<br>
「p」から始まって、3つの音から構成されるすべての語を探し出し、見つかった単語を最初と最後の音でグループ分けをしている。<br>

```python
p3 = [(pron[0] + '-' + pron[2], word)
             for (word, pron) in entries
             if pron[0] == 'P' and len(pron) == 3]
cfd = nltk.ConditionalFreqDist(p3)
for template in cfd.conditions():
    if len(cfd[template]) > 10:
        words = cfd[template].keys()
        wordlist = ' '.join(words)
        print(template, wordlist[:70]+"...")
```

### <font color='blue'>WordNet</font>

#### <font color='blue'>語義と同義語</font>

以下の文は'motorcar'と'automobile'が同義語なので、文の意味はほぼ変化しない<br>
a. Benz is credited with the invention of the motorcar.<br>
b. Benz is credited with the invention of the automobile.<br>
したがって、'motorcar'と'automobile'が同義語だと結論づけられる。

WordNetを用いてそのことを確かめる。<br>

```python
from nltk.corpus import wordnet as wn
wn.synsets('motorcar')
```

「motorcar」には、意味は 1 つしかなく、それは「car」の最初の名詞の語義を示す car.n.01 とし て識別<br>
car.n.01 というエントリは、synset あるいは同義語集合と呼ばれる。<br>

```python
wn.synset('car.n.01').lemma_names()
```

各単語の同義語集合にはいくつかの意味が含まれている。<br>
たとえば「car」には、列車の車両、ゴン ドラ、またはエレベーターの昇降機などの意味がある。<br>

同義語集合には、その単語の定義文と、いくつかの例 文が含まれている。<br>

```python
wn.synset('car.n.01').definition()
```

```python
wn.synset('car.n.01').examples()
```

同義語集合に含まれる「単語」はプログラムで利用する際にも非常に役立つ。<br>
曖昧性を避けるため、文中の単語はそれぞれ、car.n.01.automobile や car.n.01.motorcar といった形で識別<br>
こうした同義語集合と単語の組み合わせが見出し語となっている。<br>

```python
# 同義語集合からす べての見出し語を抽出
wn.synset('car.n.01').lemmas()
```

```python
# 特定の見出し語を調べる
wn.lemma('car.n.01.automobile')
```

```python
# 見出し語に対応した同義語集合を取得
wn.lemma('car.n.01.automobile').synset()
```

```python
# 見出し語の「名前」を取得
wn.lemma('car.n.01.automobile').name()
```

「automobile」や「motorcar」は、1 つの同義語集合しか持たない曖昧ではない単語<br>
「car」は 5 つの同義語集合を持つ曖昧な単語<br>

```python
wn.synsets('car')
```

```python
for synset in wn.synsets('car'):
    print(synset.lemma_names())
```

単語「car」のすべての同義語集合にアクセスできる簡便な方法が用意されている。

```python
wn.lemmas('car')
```

```python
wn.lemmas('dish')
```

### <font color='blue'>WordNetの階層構造</font>

WordNet の同義語集合は抽象概念に対応しており、対応する英語の単語がいつも存在するわけではない。<br>
最上位語 : 「実体」、「状態」、「事象」<br>
そのほかの概念 : 高燃費車」や「ハッチバック」のように、より具体的<br>

```python
Image(filename='IMG/2-0.png', width=600)
```

```python
motorcar = wn.synset('car.n.01')
```

```python
types_of_motorcar = motorcar.hyponyms()
```

```python
types_of_motorcar[26]
```

```python
motorcar.hypernyms()
```

Wordnetについては階層を上がることで、より上位の概念を手に入れることもできる。<br>
概念から下位にある 概念に至る経路が複数ある可能性もある。<br>
これはその概念が複数の上位概念を持っている。<br>
たとえばentity.n.01という概念からcar.n.01に至る経路は2つ<br>
これは wheeled_vehicle.n.01 が乗り物と容器という 2 つの概念に分類可能によること<br>

```python
paths = motorcar.hypernym_paths()
```

```python
# 2つの道順があることが分かる。
len(paths)
```

```python
# 容器
[synset.name for synset in paths[0]]
```

```python
# 乗り物
[synset.name for synset in paths[1]]
```

```python
# 同義語集合のもっとも一般的な上位語(すなわち最上位語)は以下のようにして得ることができる。
motorcar.root_hypernyms()
```

上記から容器と乗り物の２つの道が存在することが分かる。<br>

### <font color='blue'>そのほかの語彙関係</font>

上位語、下位語という概念は語彙関係と呼ばれる。<br>
これは、1 つの同義語集合からほかの同義語集合への関係を表すもの<br>

WordNetにおけるもう１つの重要な関係の表し方は以下のよう。<br>
メロニム : ある要素からそれらを含む構成要素への関係<br>
ホロニム : 構成要素とそれを含む要素との関係<br>

```python
# 幹や枝は木の一部分である。木という構成要素のそれらの部分は以下のcodeで取得可能
wn.synset('tree.n.01').part_meronyms()
```

```python
# 木を材質的に見ると以下のよう
wn.synset('tree.n.01').substance_meronyms()
```

```python
# 木が集まると以下のよう
wn.synset('tree.n.01').member_holonyms()
```

mintという単語について考えて見る<br>
この単語にはいくつかの強くたがいに関連した語義を持っている<br>
mint.n.04はmint.n.02の一部分であり、また mint.n.04はmint.n.05を作る材料でもある<br>

```python
# まずは名前と語義について見る
for synset in wn.synsets('mint', wn.NOUN):
    print(str(synset.name) + ':', str(synset.definition))
```

```python
# mint.n.04が集まると以下のよう
wn.synset('mint.n.04').part_holonyms()
```

```python
# mint.n.04の材質を見ると以下のよう
wn.synset('mint.n.04').substance_holonyms()
```

### <font color='blue'>意味類似性</font>

各同義語集合が、entity.n.01のような最上位語へリンクする1つ以上の上位語への経路を持っている。<br>
同じ最上位語へリンクしている2つの同義語集合が、いくつかの共通した上位語を持っている可能性もある。<br>
もしも２つの同義語集合が非常に具体的な上位語を共有しているのなら、それは非常に密接な関係を持っている。<br>

```python
right = wn.synset('right_whale.n.01')
```

```python
orca = wn.synset('orca.n.01')
```

```python
minke = wn.synset('minke_whale.n.01')
```

```python
tortoise = wn.synset('tortoise.n.01')
```

```python
novel = wn.synset('novel.n.01')
```

lowest_common_hypernyms()で共通の上位集合のうち最も下位の集合を出す。<br>

```python
right.lowest_common_hypernyms(minke)
```

```python
right.lowest_common_hypernyms(orca)
```

```python
right.lowest_common_hypernyms(tortoise)
```

```python
right.lowest_common_hypernyms(novel)
```

類似度の尺度はこの考え方を組み込み、WordNet同義語集合の集合を利用して定義される。<br>

path_similarityは、上位語階層の概念への最短の距離の長さに基づいた0から１の範囲を返す。<br>
もしも経路が見つからなかった場合は、-1を返す。<br>
例として、ある同義語集合と自分自身とを比較した場合は1になる。<br>
"クジラ"から"ミンクジラ", "シャチ", "カメ", "小説"という単語までの類似度スコアを考えてみる。<br>

```python
right.path_similarity(minke)
```

```python
right.path_similarity(orca)
```

```python
right.path_similarity(tortoise)
```

```python
right.path_similarity(novel)
```
