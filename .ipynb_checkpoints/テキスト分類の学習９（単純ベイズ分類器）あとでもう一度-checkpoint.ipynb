### <font color='blue'>単純ベイズ分類器</font>

詳しい数学的な手順については、言語処理のための機械学習入門を参照にする。<br>

単純ベイズ分類器は、与えられた入力値に対して、どのようなラベルがつけられるのかという時に、全ての素性が影響力を持つ。<br>

### <font color='blue'>ベイズ分類器の手順</font>

<ol>
<li>入力値のラベルを計算する為に、まずそれぞれのラベルの事前確率を計算</li>
<li>事前確率はそれぞれの各ラベルの出現頻度を調べることで得られる</li>
<li>この事前確率に各ラベルへの貢献度が組み合わされ、それぞれのラベルの尤度が推定される</li>
<li>最終的にもっとも高い尤度と判断されたラベルが、入力値に割り当てられる</li>
</ol>
