### <font color='blue'>各テキストについて他の情報を表示する</font>

#### <font color='blue'>それぞれのテキストに対して、単語の平均長、文の平均長、各単語がテキスト中に現れた回数の平均(語彙多様性スコア)</font>

```python
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid))
    num_vocab = len(set([w.lower() for w in gutenberg.words(fileid)]))
    print (int(num_chars/num_words), int(num_words/num_sents), int(num_words/num_vocab), fileid)
```

#### <font color='blue'>raw() 関数は何の言語処理も行わず、ただファイルのコンテ ンツを返す</font>
#### <font color='blue'>たとえば len(gutenberg.raw('blake-poems.txt')) は、単語間の空白 文字を含めた「文字」数を結果として返す。</font>

```python
len(gutenberg.raw('blake-poems.txt'))
```

#### <font color='blue'>法助動詞の用法から、ジャンルにおける違いを比較</font>

```python
from nltk.corpus import brown

categories = brown.categories()
del categories[2]
for c in categories:
    print(c)
    news_text = brown.words(categories = c)
    fdist = nltk.FreqDist([w.lower() for w in news_text])
    modals = ['can', 'could', 'may', 'might', 'must', 'will']
    for m in modals:
        print(m + ':', fdist[m])
    print('')
```

#### <font color='blue'>条件付き頻度分布のサポート機能を用いる</font>

下の結果から<br>
ロマンスのジャンルにおいてもっとも頻繁に現れる法助動詞が「could」なのに対し、ニュースの 場合は「will」であることがわかる。

```python
cfd = nltk.ConditionalFreqDist(
            (genre, word)
            for genre in categories
            for word in brown.words(categories=genre)
)
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditons=genres, samples=modals)
```
