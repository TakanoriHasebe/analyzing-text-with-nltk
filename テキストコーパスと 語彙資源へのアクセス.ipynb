#### <font color='blue'>それぞれのテキストに対して、単語の平均長、文の平均長、各単語がテキスト中に現れた回数の平均(語彙多様性スコア)</font>

```python
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid))
    num_vocab = len(set([w.lower() for w in gutenberg.words(fileid)]))
    print (int(num_chars/num_words), int(num_words/num_sents), int(num_words/num_vocab), fileid)
```

#### <font color='blue'>法助動詞の用法から、ジャンルにおける違いを比較</font>

```python
from nltk.corpus import brown

categories = brown.categories()
del categories[2]
for c in categories:
    print(c)
    news_text = brown.words(categories = c)
    fdist = nltk.FreqDist([w.lower() for w in news_text])
    modals = ['can', 'could', 'may', 'might', 'must', 'will']
    for m in modals:
        print(m + ':', fdist[m])
    print('')
```

#### <font color='blue'>条件付き頻度分布のサポート機能を用いる</font>

下の結果から<br>
ロマンスのジャンルにおいてもっとも頻繁に現れる法助動詞が「could」なのに対し、ニュースの 場合は「will」であることがわかる。

```python
cfd = nltk.ConditionalFreqDist(
            (genre, word)
            for genre in categories
            for word in brown.words(categories=genre)
)
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditons=genres, samples=modals)
```

### <font color='blue'>語彙資源</font>

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab.difference(english_vocab)
    return sorted(unusual)
```

```python
unusual_words(nltk.corpus.gutenberg.words('austen-sense.txt'))
```

```python
unusual_words(nltk.corpus.nps_chat.words())
```

### <font color='blue'>ストップワード</font>

高頻度で登場する言葉をストップワードという。

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

あるテキスト中にストップワードが含まれない単語の割合を計算する。

```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / len(text)

content_fraction(nltk.corpus.reuters.words())
```

上記のようにストップワードを考慮すると、テキスト中の３分の１を無視できるようになった。

### <font color='blue'>発音辞書</font>

単語だけでなくそれぞれの単語の属性情報を各行に保持した表(スプレッドシート)データ<br>
NLTK は、音声合成装置で利用するために設計された、発音辞書(CMU Pronouncing Dictionary)を含んでいる。<br>

```python
entries = nltk.corpus.cmudict.entries()
```

```python
len(entries)
```

```python
# explorersには２つの発声があるということがわかる。
for entry in entries[39943:39951]:
    print(entry)
```

上記を見ると各エントリは２つの部分から構成されている。<br>
少し複雑なfor文とプログラムにより、詳しく見ることができる。<br>

```python
for word, pron in entries:
    if len(pron) == 3:
        ph1, ph2, ph3 = pron
        if ph1 == 'P' and ph3 == 'T':
            print(word, ph2)
```

上記のプログラムは語彙目録を走査して、3つの音素からなる発音をもつエントリを探す。<br>

下記のプログラムは"nicks"のような発音で終わる単語を全て見つけ出す。<br>

```python
syllable = ['N', 'IH0', 'K', 'S']
[word for word, pron in entries if pron[-4:] == syllable]
```

上記から１つの発音が複数の表記で記される場合があるということに気づく。<br>

発音と表記の他のミスマッチの例を探して見る。<br>

以下は最後がmnで終わり、どちらかしか発音しない。<br>

```python
[w for w, pron in entries if pron[-1] == 'M' and w[-1] == 'n']
```

w[0]が'n'でなく、pron[0]が'N'で、重複を含まない。<br>

```python
sorted(set(w[:2] for w, pron in entries if pron[0] == 'N' and w[0] != 'n'))
```

語彙目録をスキャンして、発音の数字を抽出<br>
特定のアクセントパターンを 持つ語を見つけ出す関数を定義<br>

```python
def stress(pron):
    return [char for phone in pron for char in phone if char.isdigit()]
[w for w, pron in entries if stress(pron) == ['0', '1', '0', '2', '0']]
```

条件付き頻度分布は、単語の対照的な組み合わせを簡単に見つけることもできる<br>
「p」から始まって、3つの音から構成されるすべての語を探し出し、見つかった単語を最初と最後の音でグループ分けをしている。<br>

```python
p3 = [(pron[0] + '-' + pron[2], word)
             for (word, pron) in entries
             if pron[0] == 'P' and len(pron) == 3]
cfd = nltk.ConditionalFreqDist(p3)
for template in cfd.conditions():
    if len(cfd[template]) > 10:
        words = cfd[template].keys()
        wordlist = ' '.join(words)
        print(template, wordlist[:70]+"...")
```

#### <font color='blue'>語義と同義語</font>

以下の文は'motorcar'と'automobile'が同義語なので、文の意味はほぼ変化しない<br>
a. Benz is credited with the invention of the motorcar.<br>
b. Benz is credited with the invention of the automobile.<br>
したがって、'motorcar'と'automobile'が同義語だと結論づけられる。

WordNetを用いてそのことを確かめる。<br>

```python
from nltk.corpus import wordnet as wn
wn.synsets('motorcar')
```

「motorcar」には、意味は 1 つしかなく、それは「car」の最初の名詞の語義を示す car.n.01 とし て識別<br>
car.n.01 というエントリは、synset あるいは同義語集合と呼ばれる。<br>

```python
wn.synset('car.n.01').lemma_names()
```

各単語の同義語集合にはいくつかの意味が含まれている。<br>
たとえば「car」には、列車の車両、ゴン ドラ、またはエレベーターの昇降機などの意味がある。<br>

同義語集合には、その単語の定義文と、いくつかの例 文が含まれている。<br>

```python
wn.synset('car.n.01').definition()
```

```python
wn.synset('car.n.01').examples()
```

同義語集合に含まれる「単語」はプログラムで利用する際にも非常に役立つ。<br>
曖昧性を避けるため、文中の単語はそれぞれ、car.n.01.automobile や car.n.01.motorcar といった形で識別<br>
こうした同義語集合と単語の組み合わせが見出し語となっている。<br>

```python
# 同義語集合からす べての見出し語を抽出
wn.synset('car.n.01').lemmas()
```

```python
# 特定の見出し語を調べる
wn.lemma('car.n.01.automobile')
```

```python
# 見出し語に対応した同義語集合を取得
wn.lemma('car.n.01.automobile').synset()
```

```python
# 見出し語の「名前」を取得
wn.lemma('car.n.01.automobile').name()
```

「automobile」や「motorcar」は、1 つの同義語集合しか持たない曖昧ではない単語<br>
「car」は 5 つの同義語集合を持つ曖昧な単語<br>

```python
wn.synsets('car')
```

```python
for synset in wn.synsets('car'):
    print(synset.lemma_names())
```

単語「car」のすべての同義語集合にアクセスできる簡便な方法が用意されている。

```python
wn.lemmas('car')
```
