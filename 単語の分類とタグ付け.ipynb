### <font color='blue'>簡易版品詞タグセット</font>

ブラウンコーパスのニュースカテゴリの単語に対して、簡単な品詞をタグ付けする。

```python
from IPython.display import Image
Image(filename='IMG/5-0.png', width=700)
```

ブラウンコーパスのニュースカテゴリで、このタグのうちどれが一番頻出するかを調べてみる。<br>

```python
from nltk.corpus import brown

brown_news_tagged = brown.tagged_words(categories='news', tagset='universal')
tag_fd = nltk.FreqDist(tag for (word, tag) in brown_news_tagged)
tag_fd.most_common()
```

### <font color='blue'>名詞</font>

どのような品詞が名詞の前にくるか。<br>

```python
word_tag_pairs = nltk.bigrams(brown_news_tagged)
```

```python
noun_preceders = [a[1] for (a, b) in word_tag_pairs if b[1] == 'NOUN']
```

```python
fdist = nltk.FreqDist(noun_preceders)
```

```python
[tag for (tag, _) in fdist.most_common()]
```

### <font color='blue'>動詞</font>

ニュースの文章の中でもっとも一般的な動詞はなんであるかを調べる。<br>
そのために、全ての動詞を頻度順に並べる。<br>

```python
wsj = nltk.corpus.treebank.tagged_words(tagset='universal')
```

```python
word_tag_fd = nltk.FreqDist(wsj)
```

```python
word_tag_fd
```

```python
[wt[0] for (wt, _) in word_tag_fd.most_common() if wt[1] == 'VERB']
```

各単語に対してタグが付いているので、指定した単語につけられたタグを頻度順に見ることができる。

```python
cfd1 = nltk.ConditionalFreqDist(wsj)
```

```python
cfd1['yield'].most_common()
```

```python
cfd1['cut'].most_common()
```

次はタグから、どのような単語が出ているのかを調べる。<br>

```python
wsj = nltk.corpus.treebank.tagged_words()
```

```python
cfd2 = nltk.ConditionalFreqDist((tag, word) for (word, tag) in wsj)
```

```python
list(cfd2['VBN'])
```

VBD(過去形)とVBN(過去分詞)の分布の違いを明確にする。<br>
VBDとVBNのどちらにもなりうる単語を探し、その周りのテキストについて見てみる。<br>

```python
[w for w in cfd1.conditions() if 'VBD' in cfd1[w] and 'VBN' in cfd1[w]]
```

```python
idx1 = wsj.index(('kicked', 'VBD'))
```

```python
wsj[idx1-4:idx1+1]
```

```python
idx2 = wsj.index(('kicked', 'VBN'))
```

```python
wsj[idx2-4:idx2+1]
```

上記のケースでは、kickedが助動詞haveの後ろに来ることがわかる。<br>
